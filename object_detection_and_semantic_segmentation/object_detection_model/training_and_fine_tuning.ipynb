## ***Training and fine tuning YOLOv11 on my annotated dataset.***

```python
from ultralytics import YOLO

model = YOLO("/kaggle/input/yolov11m/pytorch/default/1/yolo11m.pt")

model.train(data="/kaggle/input/dataset-config-dir/dataset_config.yaml", batch=8, epochs=100, device=0)
```

```python
tuned_model = YOLO("/kaggle/working/runs/detect/train3/weights/best.pt")

tuned_model.predict("/kaggle/input/pallet-dataset/dataset/test/images/1211768-1132_jpg.rf.3b5d8745b10b9cc8ba12710c1226cbb6.jpg", show=True, save=True)
```

## ***Performace Evaluation***

```python
import matplotlib.pyplot as plt
from PIL import Image

plt.figure(figsize=(10, 20))

plt.subplot(3, 1, 1)
plt.imshow(Image.open("./runs/detect/train3/F1_curve.png"))
plt.axis('off')
plt.title("F1 Curve")

plt.subplot(3, 1, 2)
plt.imshow(Image.open("./runs/detect/train3/PR_curve.png"))
plt.axis('off')
plt.title("PR Curve")

plt.subplot(3, 1, 3)
plt.imshow(Image.open("./runs/detect/train3/results.png"))
plt.axis('off')
plt.title("Results")

plt.tight_layout()
plt.show()
```
